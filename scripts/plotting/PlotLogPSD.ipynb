```python
import ROOT
import math
```

```python
histFile = ROOT.TFile.Open("/tmp/locust-tutorial/output/basic.root","READ")
```

```python
dataHisto = histFile.Get("histPSD_0_0")
```

```python
for i in range(dataHisto.GetXaxis().GetNbins()):
    dataHisto.SetBinContent(i+1, math.log10(dataHisto.GetBinContent(i+1)))
```

```python
c = ROOT.TCanvas("myCanvasName","The Canvas Title",800,600)
```

```python
c.cd()
```

```python
dataHisto.GetYaxis().SetTitle("log10 (Power Spectral Density (W/Hz))")
dataHisto.Draw()
```

```python
c.Draw()
```
